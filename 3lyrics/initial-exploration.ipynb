```python
import pandas as pd
import re
import string
import numpy as np


# Visualization
import seaborn as sns
from matplotlib import pyplot as plt

# General Scikit learn
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

# Scikit Learn models
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn import linear_model

from pickle import dump
```

```python
songList = pd.read_csv('../data/3lyricsSongList.csv')
```

```python
songList.groupby(['genre']).count()['lyrics']/songList.groupby(['genre']).count()['songName']
```

```python
songList = songList[songList['genre'] != 'Dance/Electronic']
```

```python
songList.groupby(['genre']).count()
```

```python
countBool = ~ pd.isna(songList['lyrics']) 
dataframesList = []
for x in songList['genre'].unique():
    dataframesList.append(songList[countBool][songList['genre']==x].iloc[0:1470,:])

songListBalanced = pd.concat(dataframesList)
```

```python
def add_to_dictionary(curr_dict, lyrics):
    for word in str(lyrics).split():
        if word not in curr_dict:
            curr_dict[word] = 0 
        curr_dict[word] += 1
    return curr_dict
```

```python
def add_to_dictionary_no_add(curr_dict, lyrics):
    for word in str(lyrics).split():
        if word in curr_dict:
            curr_dict[word] += 1
    return curr_dict
```

```python
def clean_lyrics(lyrics):
    if(not pd.isna(lyrics)):
        lyrics = re.sub("[\[].*?[\]]", "", lyrics).replace('\n\n',' ').replace('\n',' ').replace('\'','').replace('\u2005',' ').lower()
        lyrics = lyrics.translate(str.maketrans('', '', string.punctuation)).strip()
    return lyrics
```

```python
songListBalanced['lyricsCleaned'] = songListBalanced.apply(lambda row: clean_lyrics(row['lyrics']), axis = 1)
```

```python
curr_dict = {}
for x in songListBalanced['lyricsCleaned']:
    curr_dict = add_to_dictionary(curr_dict, x)
curr_dict_sorted = {}
for (key, value) in sorted(curr_dict.items(), key=lambda x: x[1],reverse=True):
    curr_dict_sorted[key] = value
curr_dict_sorted.keys()
```

```python
empty_dict_sorted = {}
for x in list(curr_dict_sorted.keys())[0:1000]:
    empty_dict_sorted[x] = 0
empty_dict_sorted
```

```python
songListBalanced['lyricsDict'] = songListBalanced.apply(lambda row: add_to_dictionary_no_add(empty_dict_sorted.copy(),row['lyricsCleaned']), axis = 1)
```

```python
def tf_calculator(curr_dict,lyrics):
    lenLyrics = len(lyrics.split())
    if(lenLyrics == 0):
        return empty_dict_sorted.copy()
    curr_dict = {k: v / lenLyrics for k, v in curr_dict.items()}
    return curr_dict
```

```python
songListBalanced['lyricsDictTF'] = songListBalanced.apply(lambda row: tf_calculator(row['lyricsDict'].copy(),row['lyricsCleaned']), axis = 1)
```

```python
idf = empty_dict_sorted.copy()
for x in idf:
    for y in songListBalanced['lyricsDict']:
        if(y[x] != 0):
            idf[x] = idf[x] + 1
numDocuments = len(songListBalanced)
idf = {k: np.log(numDocuments / v) for k, v in idf.items()}
idf
```

```python
# save the model
dump(idf, open('3lyricsIDF.pkl', 'wb'))
dump(empty_dict_sorted, open('3lyricsEmptyDict.pkl', 'wb'))
```

```python
def tf_idf_calculator(curr_dict):
    return {k: curr_dict[k]*idf[k] for k in idf}
```

```python
songListBalanced['lyricsDictTFIDF'] = songListBalanced.apply(lambda row: tf_idf_calculator(row['lyricsDictTF'].copy()), axis = 1)
```

# Choosing Vocabulary

```python
sumDictTFIDF = []
curIndex = -1
for x in songList['genre'].unique():
    print(x)
    sumDictTFIDF.append(empty_dict_sorted.copy())
    curIndex = curIndex + 1
    for index, row in songListBalanced.iterrows():
        if(x == row['genre']):
            sumDictTFIDF[curIndex] = {k: row['lyricsDictTFIDF'][k] + sumDictTFIDF[curIndex][k] for k in empty_dict_sorted.copy()} 
sumDictTFIDF[0]
```

```python
TFIDFdata = {
    'Hip-Hop' : list(sumDictTFIDF[0].values()),
    'Pop' : list(sumDictTFIDF[1].values()),
    'Country' : list(sumDictTFIDF[2].values()),
    'R&B': list(sumDictTFIDF[3].values())
}
TFIDFdf = pd.DataFrame(TFIDFdata, index=list(empty_dict_sorted.keys()))
pd.set_option('display.max_rows', None)
TFIDFdf
```

```python
X = []
y = []
for index, row in songListBalanced.iterrows():
    X.append(list(row['lyricsDictTFIDF'].values()))
    y.append(row['genre'])
```

```python
Xnp = np.asarray(X)
ynp = np.asarray(y)
X_train, X_test, y_train, y_test = train_test_split(Xnp, ynp, test_size=0.2, random_state=0)
X_training, X_validation, y_training, y_validation = train_test_split(X_train, y_train, test_size=0.25, random_state=0)
```

```python
def heatmap(pred_vals, actual_vals):
    data = {'y_Actual':    actual_vals,
                'y_Predicted': pred_vals
                }
    df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
    confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
    confusion_matrix = confusion_matrix / confusion_matrix.sum(axis=1)*100
    sns.heatmap(confusion_matrix, annot=True,fmt='.3g')
    plt.tight_layout()
    plt.show()
```

```python
def scoreDataGraph(scoreData):
    print(scoreData.iloc[scoreData['validation_score'].idxmax,:])
    g = sns.regplot(x='num_steps',y='crossval_score',data=scoreData, scatter_kws={"s": 50}, order=2, ci=None, label="Cross Validation")
    g = sns.regplot(x='num_steps',y='validation_score',data=scoreData, scatter_kws={"s": 50}, order=2, ci=None, label="Testing")
    g = sns.regplot(x='num_steps',y='training_score',data=scoreData, scatter_kws={"s": 50}, order=2, ci=None, label="Training")
    g.set_ylabel("Score")
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show(g)
```

# Gaussian 

```python
scoreData = pd.DataFrame(columns=['num_steps','crossval_score','validation_score','training_score'])

for x in range(1,1000):
    gnb = GaussianNB()
    scores = cross_val_score(gnb, X_training[:,:x], y_training, cv=5)
    gnb.fit(X_training[:,:x], y_training)
    scoreData.loc[len(scoreData)] = [x, scores.mean() , gnb.score(X_validation[:,:x], y_validation), gnb.score(X_train[:,:x], y_train)]

scoreDataGraph(scoreData)
```

```python
gnb = GaussianNB()
gnb.fit(X_training[:,:524], y_train)
heatmap(gnb.predict(X_validation[:,:524]))
```

# Random Forest Classifier

```python
scoreData = pd.DataFrame(columns=['num_steps','crossval_score','validation_score','training_score'])
param_grid = [
    {'max_depth':[2,3,4,5,6,7,8,9,10],'random_state':[0]}
]
for x in range(1,500,30):
    print(x)
    clf = RandomForestClassifier()
    grid_search = GridSearchCV(clf, param_grid, cv=5)
    grid_search.fit(X_training[:,:x], y_training)
    
    scoreData.loc[len(scoreData)] = [x, grid_search.best_score_ , grid_search.best_estimator_.score(X_validation[:,:x], y_validation), grid_search.best_estimator_.score(X_training[:,:x], y_training)]


scoreDataGraph(scoreData)

```
